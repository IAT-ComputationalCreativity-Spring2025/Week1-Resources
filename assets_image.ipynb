# PIL (Pillow) Image Processing Tutorial

## Introduction
This tutorial will cover the basics of image processing using the Python Imaging Library (PIL), now known as Pillow. We'll explore various image manipulation techniques.


## Importing and Basic Operations

```python
from PIL import Image, ImageFilter, ImageEnhance
import os

# Open an image
image_path = 'image_inputs/sample_image.png'  # Replace with your image path
img = Image.open(image_path)

# NOTE: If you are using google colab, you will need to use the "files" tab on the left and upload your source.
# It will be directly in the working directory, so the path will simply be "sample_image.png"

# Display image information
print("Image Format:", img.format)
print("Image Size:", img.size)
print("Image Mode:", img.mode)

display(img)
```

## Basic Image Transformations

```python
# Resize an image
img = img.resize((256, 256))
display(img)

# Rotate an image
rotated_img = img.rotate(45)
display(rotated_img)

# Convert image to different modes
grayscale_img = img.convert('L')
display(grayscale_img)
```

## Image Filtering

```python
# Apply built-in filters
blurred_img = img.filter(ImageFilter.BLUR)
display(blurred_img)

sharp_img = img.filter(ImageFilter.SHARPEN)
display(sharp_img)

edge_enhanced_img = img.filter(ImageFilter.EDGE_ENHANCE)
display(edge_enhanced_img)

# Create a custom blur
custom_blur = ImageFilter.GaussianBlur(radius=10)
gaussian_blurred = img.filter(custom_blur)
display(gaussian_blurred)
```

## Image Enhancement

```python
# Adjust brightness
brightness_enhancer = ImageEnhance.Brightness(img)
brightened_img = brightness_enhancer.enhance(5)
display(brightened_img)

# Adjust contrast
contrast_enhancer = ImageEnhance.Contrast(img)
high_contrast_img = contrast_enhancer.enhance(5)
display(high_contrast_img)

# Adjust color
color_enhancer = ImageEnhance.Color(img)
saturated_img = color_enhancer.enhance(5)
display(saturated_img)
```

## Cropping and Copying

```python
# Crop an image
# Syntax: img.crop((left, upper, right, lower))
cropped_img = img.crop((100, 100, 300, 300))
display(cropped_img)


# Paste one image onto another
background = Image.new('RGB', (500, 500), color='white')
background.paste(cropped_img, (50, 50))
display(background)
```

## Saving Images

```python
# Create a directory for outputs if it doesn't exist
os.makedirs('image_outputs', exist_ok=True)

# Save images in different formats
img.save('image_outputs/output_image.png')
img.save('image_outputs/output_image.jpg', quality=95)

# NOTE: If you are using google colab, these files will be accessible in the files tab. You will have to download them to save
```

## Image Composition and Blending

```python
overlay = Image.new('RGBA', img.size, (255, 255, 255, 128))
blended_img = Image.alpha_composite(img.convert('RGBA'), overlay)
```

## Working with Image Modes

```python
# Convert between color modes
rgb_img = img.convert('RGB')
cmyk_img = img.convert('CMYK')
```

## Advanced: Image Drawing

```python
from PIL import ImageDraw, ImageFont

# Create a drawing context
draw_img = img.copy()
draw = ImageDraw.Draw(draw_img)

# Draw a rectangle
draw.rectangle([10, 10, 100, 100], outline='red', width=3)
display(draw_img)

# Draw text
draw.text((150, 10), "Hello, PIL!", fill='blue')
display(draw_img)
```

## Best Practices and Tips
1. Always close images after processing
2. Use context managers for file handling
3. Be mindful of memory when working with large images
4. Use appropriate image modes for your use case
